```python
import pandas as pd
import spacy
import nltk
```

```python
df_chunk = pd.read_csv('data-Project5/books.csv', iterator=False, chunksize=1000)
```

```python
df_chunk.get_chunk()
```

```python
def chunk_preprocessing(chunk, cols):
    chunk = chunk[cols]
    return pd.DataFrame(chunk)
```

```python
cols = ['review_id', 'review_date', 'product_title']
df = chunk_preprocessing(df_chunk.get_chunk(), cols)
df.head()
```

```python
chunk_list = []
for chunk in df_chunk:  
    pp_chunk = chunk_preprocessing(chunk)
    chunk_list.append(pp_chunk)
df_concat = pd.concat(chunk_list)
```
